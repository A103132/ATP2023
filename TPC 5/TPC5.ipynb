```python
### TPC5

sala1 = (150, [], "Barbie")
sala2 = (200, [], "Jumanji")
cinema1 = [sala1, sala2]

Menu = print('''Bem vindo ao cinema GG\nMenu:\n1. Filmes em exibição\n2. Disponibilidade de Lugares: Insere o lugar que pretendes ocupar\n3. Lugares vendidos\n4. Filme em exibição + Lugares disponiveis\n5. Inserir Sala ao Cinema\n0. Sair''')
opc = int(input("Selecione a opção do menu que pretende efetuar! Se quiser encerrar o menu, selecione a opção 0."))


def listar(cinema):
    print ("------- FILMES EM EXIBIÇÃO -------")
    lista = []
    for sala in cinema:
        print("Nome: ",sala[2],"    |    ", "Nº de lugares: ", sala[0])

def disponivel(cinema, filme, lugar):
    cond = False
    for sala in cinema:
        if filme == sala[2]:
            if lugar not in sala[1]:
                cond = True

        return cond

def venderbilhete (cinema, filme, lugar):
    for sala in cinema:
        if filme==sala[2]:
            if lugar not in sala[1]:
                sala[1].append(lugar)


def listardisponibilidades(cinema):
    for sala in cinema:
        print("O filme :", sala[2] , " tem", sala[0], "lugares disponiveis e ", sala[1], " lugares ocupados")
    return cinema


def inserirsala(cinema, sala):
    cinema.append(sala)
    return cinema


while opc != 0:

    if opc == 1:
        listar(cinema1)
        opc = int(input("Selecione a opção do menu que pretende efetuar! Se quiser encerrar o menu, selecione a opção 0."))
    

    if opc == 2:
        cinema = str(input("Escolha o cinema"))
        filme = str(input("Qual é o filme que quer ver"))
        lugar = str(input("Escolha o lugar"))

        if disponivel(cinema1, filme, lugar) == False :
            print("O lugar selecionado está disponível")

        else:
            print("O lugar selecionado não está disponível")
        opc = int(input("Selecione a opção do menu que pretende efetuar! Se quiser encerrar o menu, selecione a opção 0."))

    if opc == 3:
        cinema = str(input("Escolha o cinema"))
        filme = str(input("Qual é o filme que quer ver?"))
        lugar = int(input("Escolha o lugar"))
        venderbilhete(cinema1, filme, lugar)
        print("O seu lugar numero "+str(lugar)+" está disponivel, por isso pode ser vendido." )
        opc = int(input("Selecione a opção do menu que pretende efetuar! Se quiser encerrar o menu, selecione a opção 0."))

 
    if opc == 4:
        listardisponibilidades(cinema1)
        opc = int(input("Selecione a opção do menu que pretende efetuar! Se quiser encerrar o menu, selecione a opção 0."))


    if opc == 5:
    
        filme = str(input("Insira o filme que estará em exibição na sala"))
        lugaresDisp = int(input("Insira o número de lugares disponiveis na sala"))
        lugaresVend = int(input("Quantos lugares já estão vendidos?"))
          
        sala3 = (lugaresDisp,lugaresVend,filme)
        inserirsala(cinema1, sala3)
        print("Adicionou o filme : " +str(filme)+ " com " +str(lugaresDisp)+ " lugares disponiveis e " +str(lugaresVend)+ " lugares já vendidos!")
        opc = int(input("Selecione a opção do menu que pretende efetuar! Se quiser encerrar o menu, selecione a opção 0."))
    

print("Muito Obrigada! Volte Sempre.")
```
